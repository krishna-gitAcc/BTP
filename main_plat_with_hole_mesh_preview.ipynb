```python
import mesh_gen
from matplotlib import pyplot as plt
mesh_obj = mesh_gen.platWithHole(4, 4, [2, 0, 7, 0, 7, 7, 0, 7, 0, 2], 0)
coord = mesh_obj.coord_array()
connectivity = mesh_obj.connectivity()
# print(coord)
# print(connectivity)
xs = [x[0] for x in coord]
ys = [x[1] for x in coord]
plt.scatter(xs, ys)
plt.show()
# Plot the nodal coordinates
plt.scatter(coord[:, 0], coord[:, 1])

# Plot the element connectivity
for conn in connectivity:
    # print(conn)
    n1, n2, n3, n4 = conn
    plt.plot([coord[n1, 0], coord[n2, 0], coord[n3, 0], coord[n4, 0], coord[n1, 0]],
             [coord[n1, 1], coord[n2, 1], coord[n3, 1], coord[n4, 1], coord[n1, 1]],
             'k-')

#Set the plot limits and axis labels
Lx = 7;
Ly = 7;
plt.xlim([0, Lx])
plt.ylim([0, Ly])
plt.xlabel('X')
plt.ylabel('Y')
```
